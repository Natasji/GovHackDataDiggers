Some functions (round coordinates)

```python
import pandas as pd
# Function to round LATITUDE and LONGITUDE to n decimal places, same as other dfs
def round_lat_lon(df, places=5):
    df['LATITUDE'] = df['LATITUDE'].round(places)
    df['LONGITUDE'] = df['LONGITUDE'].round(places)
```

# Use accident.csv as base

```python
df = pd.read_csv('ACCIDENT.csv')

# Extract neccessay columns
df = df[['ACCIDENT_NO', 'NODE_ID', 'ACCIDENT_DATE', 'ACCIDENT_TIME', 'ACCIDENT_TYPE', 'DAY_OF_WEEK', 'SEVERITY', 'ROAD_GEOMETRY', 'SPEED_ZONE']]

df
```

```python
dataset = {
    'accident location': {'file': 'ACCIDENT_LOCATION.csv', 'columns': ['ACCIDENT_NO', 'NODE_ID', 'ROAD_TYPE', 'ROAD_TYPE_INT'], 'id': ['ACCIDENT_NO', 'NODE_ID']},
    'atmospheric': {'file': 'ATMOSPHERIC_COND.csv', 'columns': ['ACCIDENT_NO', 'ATMOSPH_COND'], 'id': ['ACCIDENT_NO']},
    'node': {'file': 'NODE.csv', 'columns': ['ACCIDENT_NO', 'NODE_ID', 'LATITUDE', 'LONGITUDE', 'POSTCODE_CRASH'], 'id': ['ACCIDENT_NO', 'NODE_ID']},
    'person': {'file': 'PERSON.csv', 'columns': ['ACCIDENT_NO', 'AGE_GROUP'], 'id': ['ACCIDENT_NO']},
    'road surface': {'file': 'ROAD_SURFACE_COND.csv', 'columns': ['ACCIDENT_NO', 'SURFACE_COND'], 'id': ['ACCIDENT_NO']},
    'sub_dca': {'file': 'SUB_DCA.csv', 'columns': ['ACCIDENT_NO', 'SUB_DCA_CODE'], 'id': ['ACCIDENT_NO']},
    'speed sign': {'file': 'Speed_Signs_August_2024_csv.csv', 'columns': ['sign_size', 'latitude', 'longitude', 'speed', 'sign_type'], 'id': ['LATITUDE', 'LONGITUDE']},
    'traffic lights': {'file': 'Traffic_Lights_csv.csv', 'columns': ['latitude', 'longitude', 'SITE_TYPE' ], 'id': ['LATITUDE', 'LONGITUDE']}
}
```

# *test view data

```python
test = "traffic lights"
temp_df = pd.read_csv(dataset[test]['file'])

# Extract neccessay columns
temp_df = temp_df[dataset[test]['columns']]

temp_df.columns = temp_df.columns.str.upper()
if 'LATITUDE' in temp_df.columns and 'LONGITUDE' in temp_df.columns:
    round_lat_lon(temp_df)

print(temp_df.head())
```

# Merge datasets

```python
for file in dataset:
    temp_df = pd.read_csv(dataset[file]['file'])
    temp_df = temp_df[dataset[file]['columns']]
    temp_df.columns = temp_df.columns.str.upper()
    if 'LATITUDE' in temp_df.columns and 'LONGITUDE' in temp_df.columns:
        round_lat_lon(temp_df)
    df = pd.merge(df, temp_df, on=dataset[file]['id'], how='left')

# Display the updated node_df DataFrame
df
```

```python
df.to_csv('combined.csv', index=False)
```
